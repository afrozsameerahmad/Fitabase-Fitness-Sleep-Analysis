```python
sleep_data["HoursSlept"] = sleep_data.loc[:,"TotalMinutesAsleep"]/60
sleep_data["TimeTakenToSleep"] = sleep_data.loc[:,"TotalTimeInBed"]-sleep_data.loc[:,"TotalMinutesAsleep"]
```

```python
df["ActivityDate"]= pd.to_datetime(df.loc[:,"ActivityDate"])
```

```python
df = df[['Id', 'ActivityDate', 'TotalSteps', 'TotalDistance',
             #'TrackerDistance',
       #'LoggedActivitiesDistance',
             'VeryActiveDistance',
       'ModeratelyActiveDistance', 'LightActiveDistance',
       #'SedentaryActiveDistance', 
             'VeryActiveMinutes', 'FairlyActiveMinutes',
       'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories']]
```

```python
df =df.rename(columns={
    "ModeratelyActiveDistance":"FairlyActiveDistance",
    "LightActiveDistance":"LightlyActiveDistance"
})
```

```python
df["Weekday"] = df.loc[:,"ActivityDate"].dt.day_name()
```

```python
df["TotalActiveMinutes"]=df.loc[:,"VeryActiveMinutes"]+df.loc[:,"FairlyActiveMinutes"]\
                    +df.loc[:,"LightlyActiveMinutes"]
df["TotalHours"]=round((df.loc[:,"VeryActiveMinutes"]+df.loc[:,"FairlyActiveMinutes"]\
                    +df.loc[:,"LightlyActiveMinutes"]+df.loc[:,"SedentaryMinutes"])/60).astype(int)
```

# Which weekday had the most activity?

```python
df.groupby("Weekday")["TotalActiveMinutes"].sum().sort_values(ascending=False).plot(kind = "bar",color = "blue")
plt.title("Total activity across weekdays")
plt.show()
```

Here we can see that Tuesday had the most activity followed by saturday then Friday had the highest activity

# What are people mostly using their watch for?

```python
minutes = df.loc[:,["VeryActiveMinutes","FairlyActiveMinutes","LightlyActiveMinutes","SedentaryMinutes"]].sum()
```

```python
plt.figure(figsize=(12,9))
plt.pie(minutes,labels=minutes.index,autopct='%1.2f%%',colors= ["Red","Yellow","Green","purple"])
plt.show()
```

```python
distance = df.loc[:,["VeryActiveDistance","FairlyActiveDistance","LightlyActiveDistance"]].sum()
```

```python
plt.figure(figsize=(12,9))
plt.pie(distance,labels=distance.index,autopct='%1.2f%%',colors=["Red","Yellow","Green"])
plt.show()
```

From the 2 pie charts above we can see that most of the people are using the watch for normal activies with low high activity time and distance
